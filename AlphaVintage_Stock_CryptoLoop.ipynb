```python
# url=f'https://api.marketstack.com/v1/eod?access_key={marketstack}&symbols=AAPL'


# r = requests.get(url)
# data = r.json()

# print(json.dumps(data, indent=4, sort_keys=True))

import requests

params = {
  'access_key': marketstack
}

tickers=['aapl','goog','msft']

historical_no_of_days=100 #set to 5 years (252 * 5 = 1260)

url=[]

json=[]

for index,symbol in enumerate(tickers):
    print(index)
    print(symbol)
    api_index=f'https://api.marketstack.com/v1/eod?access_key={marketstack}&symbols={symbol}&limit={historical_no_of_days}'
    url.append(api_index)

# url
for link in url:
    api_result = requests.get(link).json()
    json.append(api_result)

# json[0]
```

```python
makedataframe=pd.DataFrame(columns=
                           [
                            'open'
                            ,'high'
                            ,'low'
                            ,'close'
                            ,'volume'
                            ,'adj_high'
                            ,'adj_low'
                            ,'adj_close'
                            ,'adj_open'
                            ,'adj_volume'
                            ,'split_factor'
                            ,'dividend'
                            ,'symbol'
                            ,'exchange'
                            ,'date'
                    ])

display(makedataframe)
# Initialize iterative variables
open = 0.0
high = 0.0
low = 0.0
close = 0.0
volume = 0.0
adj_high = 0.0
adj_low = 0.0
adj_close = 0.0
adj_open = 0.0
adj_volume = 0.0
split_factor = 0.0
dividend = 0.0
symbol=''
exchange=''
date=''

for data in json:
    for eoddata in data['data']:
            open =     eoddata['open']
            high =     eoddata['high']
            low  =     eoddata['low']
            close =    eoddata['close']
            volume=    eoddata['volume']
            adj_high = eoddata['adj_high']
            adj_low =  eoddata['adj_low']
            adj_close= eoddata['adj_close']
            adj_open = eoddata['adj_open']
            adj_volume =   eoddata['adj_volume']
            split_factor = eoddata['split_factor']
            dividend =     eoddata['dividend']
            symbol=        eoddata['symbol']
            exchange=      eoddata['exchange']
            date=          eoddata['date']
                
            makedataframe=makedataframe.append(
                                                {'open'      : open
                                                ,'high'      : high
                                                ,'low'       : low
                                                ,'close'     : close
                                                ,'volume'     : volume
                                                ,'adj_high'  : adj_high 
                                                ,'adj_low'   : adj_low
                                                ,'adj_close' : adj_close
                                                ,'adj_open'  : adj_open
                                                ,'adj_volume': adj_volume
                                                ,'split_factor'  : split_factor
                                                ,'dividend'      : dividend
                                                ,'symbol'        : symbol
                                                ,'exchange'      : exchange
                                                ,'date'          : date
                                                },
                ignore_index=True
            )

# # makedataframe['date']=pd.to_datetime(makedataframe['date'])
display(makedataframe.info())
display(makedataframe.loc[makedataframe['symbol'] == tickers[0].upper()].head())
display(makedataframe.loc[makedataframe['symbol'] == tickers[0].upper()].tail())
```
